### Embedding, sözlük anlamların sayısal değerleri

```python
import torch

embedding = torch.nn.Embedding(num_embeddings=64, embedding_dim=4)
```

```python
embedding(torch.tensor(token_ids))
```

```python
example = "the capital of united states"
example_tokens = tokenizer.encode("the capital of united states")
example_tokens
```

```python
example_meaning = embedding(torch.tensor(example_tokens))
example_meaning
```

```python
example_meaning.shape
```

```python
import plotly.graph_objects as go
import plotly.offline

def plot_dots(sentences_data, title, dims=[0, 1, 2]):
    data = [
      go.Scatter3d(
        x=sentence_data["words"][:, dims[0]],
        y=sentence_data["words"][:, dims[1]],
        z=sentence_data["words"][:, dims[2]],
        mode="markers+text",
        marker=dict(
          size=6,
          color=sentence_data["color"],
        ),
        text=sentence_data["labels"],
        hoverinfo="text",
      ) for sentence_data in sentences_data
    ]

    layout = go.Layout(
      scene=dict(
        xaxis_title="Sertlik",
        yaxis_title="Parlaklık",
        zaxis_title="Kırmızılık",
      ),
      title=title,
    )

    fig = go.Figure(data=data, layout=layout)
    plotly.offline.iplot(fig)
```

```python
tokenizer.tokenizer(example)
```

```python
sentences = [
  {
    "words": example_meaning.detach().numpy(),
    "labels": tokenizer.tokenizer(example),
    "color": "red",
  },
]
plot_dots(sentences, "Sözlük V1")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, Gemma3ForCausalLM

gemma_tokenizer = AutoTokenizer.from_pretrained("google/gemma-3-1b-it")
gemma_model = Gemma3ForCausalLM.from_pretrained("google/gemma-3-1b-it")
```

```python
gemma_model
```

```python
gemma_model.model.embed_tokens
```

```python
gemma_tokens = gemma_tokenizer.encode("the capital of united states")
gemma_tokens
```

```python
gemma_meaning = gemma_model.model.embed_tokens(torch.tensor(gemma_tokens))
gemma_meaning.shape
```

```python
gemma_meaning
```

```python
gemma_tokenizer.tokenize("the capital of united states")
```

```python
sentences = [
  {
    "words": gemma_meaning.detach().cpu().float().numpy(),
    "labels": gemma_tokenizer.tokenize("the capital of united states"),
    "color": "red",
  },
]
plot_dots(sentences, "Gemma", dims=[20,21,22])
```

### cümledeki yerine göre embedding

```python
depend_sentence = "the capital of united states and the capital of france"
```

```python
depend_sentence_token_ids = tokenizer.encode("the capital of united states")
depend_sentence_tokens = tokenizer.tokenizer(depend_sentence)
```

```python
depend_sentence_meaning = embedding(torch.tensor(depend_sentence_token_ids))

sentences = [
  {
    "words": depend_sentence_meaning.detach().numpy(),
    "labels": depend_sentence_tokens,
    "color": "red",
  },
]
plot_dots(sentences, "Depends on Sentence")
```

```python
gemma_depend_sentence_token_ids = gemma_tokenizer.encode(depend_sentence)
gemma_depend_sentence_tokens = gemma_tokenizer.tokenize(depend_sentence)
gemma_depend_sentence_meaning = gemma_model.model.embed_tokens(torch.tensor(gemma_depend_sentence_token_ids))
```

```python
sentences = [
  {
    "words": gemma_depend_sentence_meaning.detach().cpu().float().numpy(),
    "labels": gemma_depend_sentence_tokens,
    "color": "red",
  },
]
plot_dots(sentences, "Depends on Sentence")
```
